```python
import matplotlib.pyplot as plt
from kvBench import *
```

```python
def time_list(is_mixed=False, is_manyRW=False, is_cold=True, is_R=True, is_W=True):
    np.random.seed(50)
    direct_lst = np.array(time_putAndget('direct', is_mixed, is_manyRW, is_cold, is_R, is_W))
    redis_lst = np.array(time_putAndget('process', is_mixed, is_manyRW, is_cold, is_R, is_W))
    anna_lst = np.array(time_putAndget('Anna', is_mixed, is_manyRW, is_cold, is_R, is_W))
    return direct_lst, redis_lst, anna_lst
```

```python
X = [['5by5' for _ in range(10)], ['100by100' for _ in range(10)], \
     ['1000by1000' for _ in range(10)], ['5000by5000' for _ in range(10)]]
X = np.array(X).flatten()
```

### Test for is_cold

```python
dl1, rl1, al1 = time_list()
```

```python
plt.figure(figsize=(12, 9))
plt.boxplot(np.log(dl1).tolist(), positions=[1,5,9,13], labels=["5by5d","100by100d","1000by1000d","5000by5000d"])
plt.boxplot(np.log(rl1).tolist(), positions=[2,6,10,14], labels=["5by5r","100by100r","1000by1000r","5000by5000r"])
plt.boxplot(np.log(al1).tolist(), positions=[3,7,11,15], labels=["5by5a","100by100a","1000by1000a","5000by5000a"])
plt.xticks(rotation=90)
plt.xlabel("different matrices in different databases")
plt.ylabel("execution time in log base")
plt.title("Execution Time when is_cold=True (box plot)")
plt.show()
```

```python
plt.figure(figsize=(12, 9))
plt.scatter(X, np.log(dl1.flatten()), label="direct")
plt.scatter(X, np.log(rl1.flatten()), label="redis")
plt.scatter(X, np.log(al1.flatten()), label="anna")
plt.xlabel("different matrices in different databases")
plt.ylabel("execution time in log base")
plt.legend()
plt.title("Execution Time when is_cold=True (scatter plot)")
#plt.show()
plt.savefig("is_coldT")
```

```python
dl2, rl2, al2 = time_list(is_cold=False)
```

```python
plt.figure(figsize=(12, 9))
plt.boxplot(np.log(dl2).tolist(), positions=[1,5,9,13], labels=["5by5d","100by100d","1000by1000d","5000by5000d"])
plt.boxplot(np.log(rl2).tolist(), positions=[2,6,10,14], labels=["5by5r","100by100r","1000by1000r","5000by5000r"])
plt.boxplot(np.log(al2).tolist(), positions=[3,7,11,15], labels=["5by5a","100by100a","1000by1000a","5000by5000a"])
plt.xticks(rotation=90)
plt.xlabel("different matrices in different databases")
plt.ylabel("execution time in log base")
plt.title("Execution Time when is_cold=False (box plot)")
plt.show()
```
